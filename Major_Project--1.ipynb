```python
import datetime as dt
import matplotlib.patches as mpatches
import numpy as np

plt.figure(figsize=(10,6))


dates = []
for year in range(2016, 2017):
    for month in range(11, 13):
        dates.append(dt.datetime(year=year, month=month, day=1))
for year in range(2017, 2018):
    for month in range(1, 8):
        dates.append(dt.datetime(year=year, month=month, day=1))

red_patch = mpatches.Patch(color='green', label='Positive opinion')
green_patch = mpatches.Patch(color='red', label='Negative opinion')
yellow_patch = mpatches.Patch(color='y', label='Neutral opinion')
plt.legend(handles=[red_patch,green_patch])
plt.xlabel('Month')
plt.ylabel('Sentiment Score')
plt.title('Positive vs Negative vs Neutral sentiments')
plt.plot(dates,pos_arr,'g')
plt.plot(dates,neu_arr,'y')
plt.plot(dates,neg_arr,'r')

print ("POSITIVE = ",sum(pos_arr)/len(pos_arr),"NEGATIVE = ",sum(neg_arr)/len(pos_arr),"NEUTRAL = ",sum(neu_arr)/len(pos_arr))


plt.show()
```
